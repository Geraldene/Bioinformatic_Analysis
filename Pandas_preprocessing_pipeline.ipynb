```python
#missing values 
def fill_missing_values(df):
    for col in df.select_dtypes(include=['int', 'float']).columns:
        val = df[col].mean()
        df[col].fillna(val, inplace=True)
    return df
```

```python
#remove duplicates 
def drop_duplicates(df, column_name):
    df=df.drop_duplicates(subset=column_name, inplace=False)
    return df
```

```python
#remove outliers
def remove_outliers(df, column_list):
    for col in column_list:
        avg = df[col].mean()
        std = df[col].std()
        low = avg - 2 * std 
        high = avg + 2 * std 
        df = df[df[col].between(low, high, inclusive=True)]
    return df
```

```python
#create pipe

df_processed = (df.pipe(fill_missing_values). 
               pipe(drop_duplicates, 'AveragePrice').
               pipe(remove_outliers, ['AveragePrice']))
```

```python
#run 
my_df = df.copy()

df_processed = (my_df.pipe(fill_missing_values). 
               pipe(drop_duplicates, 'AveragePrice').
               pipe(remove_outliers, ['AveragePrice']))
```
